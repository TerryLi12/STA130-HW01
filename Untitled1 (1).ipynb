```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset (replace 'path_to_your_csv' with your file path)
data = pd.read_csv('https://raw.githubusercontent.com/pointOfive/stat130chat130/refs/heads/main/CP/CSCS_data_anon.csv')

# Visualization 1: Histogram for Life Satisfaction
plt.figure(figsize=(8, 5))
sns.histplot(data['WELLNESS_life_satisfaction'], bins=10, kde=False)
plt.title('Distribution of WELLNESS_life_satisfaction')
plt.xlabel('Life Satisfaction (1-10)')
plt.ylabel('Frequency')
plt.show()

# Visualization 2: Bar Plot for Community Activities
plt.figure(figsize=(8, 5))
sns.countplot(
    y=data['CONNECTION_activities_community_p3m'],
    order=data['CONNECTION_activities_community_p3m'].value_counts().index
)
plt.title('Frequency of Participation in Community Activities')
plt.xlabel('Frequency')
plt.ylabel('Community Activities Responses')
plt.show()

# Visualization 3: Boxplot for Work Hours per Week
plt.figure(figsize=(8, 5))
sns.boxplot(x=data['WORK_hours_per_week'])
plt.title('Boxplot of WORK_hours_per_week')
plt.xlabel('Work Hours per Week')
plt.show()

# Visualization 4: Bar Plot for Emotional Support Seeking
plt.figure(figsize=(8, 5))
sns.countplot(
    y=data['PSYCH_cope_60_inventory_get_emotional_support'],
    order=data['PSYCH_cope_60_inventory_get_emotional_support'].value_counts().index
)
plt.title('Frequency of Seeking Emotional Support')
plt.xlabel('Frequency')
plt.ylabel('Emotional Support Responses')
plt.show()
```

```python
# Import necessary libraries
import pandas as pd

# Select numeric variables for analysis
numeric_vars = ['WORK_hours_per_week', 'WELLNESS_life_satisfaction']

# Calculate summary statistics
summary_stats = data[numeric_vars].describe()

# Calculate IQR for each numeric variable
iqr = data[numeric_vars].quantile(0.75) - data[numeric_vars].quantile(0.25)

# Add IQR to the summary statistics for easier reference
summary_stats.loc['IQR'] = iqr

# Display summary statistics
print(summary_stats)
```

```python
# Handle missing data systematically by verifying columns individually

# Reload the full dataset to start fresh
full_data_cleaned = data

# Drop rows with missing values for critical numeric variables
critical_numeric_vars = ['WORK_hours_per_week', 'WELLNESS_life_satisfaction']
full_data_cleaned = full_data_cleaned.dropna(subset=critical_numeric_vars)

# Impute missing values for numeric variables with the median
numeric_vars_full = full_data_cleaned.select_dtypes(include=['float64', 'int64']).columns
for var in numeric_vars_full:
    try:
        full_data_cleaned[var].fillna(full_data_cleaned[var].median(), inplace=True)
    except Exception as e:
        print(f"Error imputing numeric column {var}: {e}")

# Impute missing values for categorical variables with the mode
categorical_vars_full = full_data_cleaned.select_dtypes(include=['object']).columns
for var in categorical_vars_full:
    try:
        if not full_data_cleaned[var].mode().empty:
            full_data_cleaned[var].fillna(full_data_cleaned[var].mode()[0], inplace=True)
    except Exception as e:
        print(f"Error imputing categorical column {var}: {e}")

# Verify no missing data remains
missing_summary_full_cleaned = full_data_cleaned.isna().sum()

# Display the cleaned dataset and missing data summary
missing_summary_full_cleaned.head(), full_data_cleaned.head()
```
